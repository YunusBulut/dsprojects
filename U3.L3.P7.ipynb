```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

# Letâ€™s make histograms for the two groups. 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
```

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
print("sample 1 mean is " + str(sample1.mean()))
print("sample 1 stdev is " + str(sample1.std()))
print("sample 2 mean is " + str(sample2.mean()))
print("sample 2 stdev is " + str(sample2.std()))
```

```python
#increase sample sizes to 1000, i'd expect the means to get closer to their true values, I wouldn't expect a meaningful 
#change in the standard deviation.
sample1a = np.random.choice(pop1, 1000, replace=True)
sample2a = np.random.choice(pop2, 1000, replace=True)

plt.hist(sample1a, alpha=0.5, label='sample 1a') 
plt.hist(sample2a, alpha=0.5, label='sample 2a') 
plt.legend(loc='upper right') 
plt.show()
print("sample 1a mean is " + str(sample1a.mean()))
print("sample 1a stdev is " + str(sample1a.std()))
print("sample 2a mean is " + str(sample2a.mean()))
print("sample 2a stdev is " + str(sample2a.std()))
```

```python
#with decreasing sample sizes to 20, I'd expect more noise in the mean and the standard deviation of both
sample1b = np.random.choice(pop1, 20, replace=True)
sample2b = np.random.choice(pop2, 20, replace=True)

plt.hist(sample1b, alpha=0.5, label='sample 1b') 
plt.hist(sample2b, alpha=0.5, label='sample 2b') 
plt.legend(loc='upper right') 
plt.show()
print("sample 1b mean is " + str(sample1b.mean()))
print("sample 1b stdev is " + str(sample1b.std()))
print("sample 2b mean is " + str(sample2b.mean()))
print("sample 2b stdev is " + str(sample2b.std()))
```

```python
pop3 = np.random.binomial(10, 0.3, 10000)


# Letâ€™s make histograms for the two groups. 

plt.hist(pop3, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
```

```python
sample3 = np.random.choice(pop1, 100, replace=True)


plt.hist(sample3, alpha=0.5, label='sample 3') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
print("sample 3 mean is " + str(sample3.mean()))
print("sample 3 stdev is " + str(sample3.std()))
print("sample 2 mean is " + str(sample2.mean()))
print("sample 2 stdev is " + str(sample2.std()))
```

```python
diff32 = sample2.mean() - sample3.mean()
#print(diff32)
size = np.array([len(sample3), len(sample2)])
sd = np.array([sample3.std(), sample2.std()])
se32 = (sum(sd**2/size))**0.5

tval32 = diff32/se32
print(tval32)
```

```python
from scipy.stats import ttest_ind
print(ttest_ind(sample3, sample2,equal_var=False))
```

```python
#by changing the distribution p values to be closer, I'd expect the t value to get closer to zero and the p value to increase.
pop4 = np.random.binomial(10, 0.4, 10000)
sample4 = np.random.choice(pop4, 100, replace=True)

plt.hist(sample3, alpha=0.5, label='sample 3') 
plt.hist(sample4, alpha=0.5, label='sample 4') 
plt.legend(loc='upper right') 
plt.show()
print("sample 3 mean is " + str(sample3.mean()))
print("sample 3 stdev is " + str(sample3.std()))
print("sample 4 mean is " + str(sample4.mean()))
print("sample 4 stdev is " + str(sample4.std()))
```

```python
diff43 = sample4.mean() - sample3.mean()
#print(diff32)
size = np.array([len(sample3), len(sample4)])
sd = np.array([sample3.std(), sample4.std()])
se43 = (sum(sd**2/size))**0.5

tval43 = diff43/se43
print(tval43)
```

```python
print(ttest_ind(sample3, sample4,equal_var=False))
```

```python
#make 2 poisson distributions that are relatively similar

pop5 = np.random.poisson(lam=1.0, size=10000)
pop6 = np.random.poisson(lam=1.3, size=10000)

plt.hist(pop5, alpha=0.5, label='Population 5') 
plt.hist(pop6, alpha=0.5, label='Population 6') 
plt.legend(loc='upper right') 
plt.show()
```

```python
#take 100 samples to test CLT
sample5 = np.random.choice(pop5, 100, replace=True)
sample6 = np.random.choice(pop6, 100, replace=True)

plt.hist(sample5, alpha=0.5, label='sample 5') 
plt.hist(sample6, alpha=0.5, label='sample 6') 
plt.legend(loc='upper right') 
plt.show()
print("sample 5 mean is " + str(sample5.mean()))
print("sample 5 stdev is " + str(sample5.std()))
print("sample 6 mean is " + str(sample6.mean()))
print("sample 6 stdev is " + str(sample6.std()))
```

```python
diff65 = sample6.mean() - sample5.mean()
size = np.array([len(sample6), len(sample5)])
sd = np.array([sample6.std(), sample5.std()])
se65 = (sum(sd**2/size))**0.5

tval65 = diff65/se65
print(tval65)
```

```python
print(ttest_ind(sample6, sample5,equal_var=False))
```

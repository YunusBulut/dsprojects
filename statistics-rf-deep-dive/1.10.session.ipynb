```python
import pandas as pd

from sklearn.datasets import load_boston, load_iris
from sklearn.ensemble import RandomForestClassifier
import statsmodels.formula.api as smf
```

# Regression Dataset for statsmodels examination

```python
boston = load_boston()
print(boston.DESCR)
```

```python
dfreg = pd.DataFrame(boston.data, columns=boston.feature_names)
dfreg['MEDV'] = boston.target
print(dfreg.shape)
dfreg.head()
```

```python
dfreg.columns
```

```python
linearmodel = 'MEDV ~ CRIM+ZN+INDUS+CHAS+NOX+RM+AGE+DIS+RAD+TAX+PTRATIO+B+LSTAT'
results = smf.ols(linearmodel, data=dfreg).fit()
results.summary()
```
